```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelBinarizer, LabelEncoder
from matplotlib import pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('//content//sample_data//Mall_Customers (1).csv')
df= pd.DataFrame(data)
df
```

```python
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df
```

```python
cluster_no = (i for i in range(1, 11))
wcss = []
for k in cluster_no:
  kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300, n_init=10, random_state=0)
  kmeans.fit(df)
  wcss.append(kmeans.inertia_)
plt.figure(figsize=(8,6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()



```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(df)

plt.figure(figsize=(8,6))
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids', marker='X')
plt.title(f'K-Means clustering (K={optimal_k})')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))

# Scatter plot for each cluster
for i in range(optimal_k):
    plt.scatter(df[y_kmeans == i]['Annual Income (k$)'], df[y_kmeans == i]['Spending Score (1-100)'], s=50, label=f'Cluster {i+1}')

# Plot centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', label='Centroids', marker='X')

plt.title(f'K-Means clustering (K={optimal_k})')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
